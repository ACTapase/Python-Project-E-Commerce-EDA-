## 4. Remove rows which have negative values in Quantity column

```python
df = ecom_df[(ecom_df['Quantity'] >= 0)]
df.info()
```

## 5. Add the columns - Month, Day and Hour for the invoice

```python
ecom_df['InvoiceDate'] = ecom_df['InvoiceDate'].astype('datetime64[ns]')
ecom_df['Year'] = ecom_df.InvoiceDate.dt.year
ecom_df['Month'] = ecom_df.InvoiceDate.dt.month
ecom_df['year_month']=ecom_df.InvoiceDate.map(lambda x: 100*x.year + x.month)
ecom_df['Day']=(ecom_df.InvoiceDate.dt.dayofweek)+1 # +1 to make Monday=1.....until Sunday=7
ecom_df['Hour'] = ecom_df.InvoiceDate.dt.hour
ecom_df['Revenue'] =ecom_df['Quantity'] * ecom_df['UnitPrice']
ecom_df.head()
```

## 13. How many orders for each country?

```python
group_country_orders = ecom_df.groupby('Country')['InvoiceNo'].count().sort_values()
del group_country_orders['United Kingdom']

# number of unique customers in each country (without UK)
plt.subplots(figsize=(15,8))
group_country_orders.plot(kind='barh', fontsize=14, color=color[0])
plt.xlabel('Number of Orders', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.title('Number of Orders for different Countries', fontsize=14)
plt.show()
```

## 14. Orders trend across months

```python
# trend plot
fig = plt.figure(figsize=(15, 7))
x1 = fig.add_subplot(111)

x1.set_xlabel('Month',fontsize=15)
x1.set_ylabel('Total Amount Spent ($)',fontsize=15)
x1.set_title('Total Amount Spent for Months',fontsize=15)

x1.bar(ecom_df.groupby(by=['year_month'])['Revenue'].sum().index.tolist()[1:],
        ecom_df.groupby(by=['year_month'])['Revenue'].sum()[1:],
        alpha=0.85,#If you want to make the graph plot more transparent, then you can make alpha less than 1, such as 0.5 or 0.25.
        label='Amount Spent by Month')

x2 = x1.twinx()# tread line

x2.set_ylabel('Percentage Change of Total Amount Spent (%)',fontsize=15)

#pct_change()Percentage change between the current and a prior element.
x2.plot(ecom_df.groupby(by=['year_month'])['Revenue'].sum().index.tolist()[1:], 
         ecom_df.groupby(by=['year_month'])['Revenue'].sum()[1:].pct_change().fillna(0)*100,
         label='Percentage change of total amount spent (%)',
         color='red')

x1.legend(loc='upper left')
x2.legend(loc='upper right')

fig.tight_layout()
plt.show()
```
